```python
from tqdm.autonotebook import tqdm
obsl=[]
n_samples=20
for i in tqdm(range(n_samples)):
    obs=env.sample_and_process_observations_from_device()
    obsl.append(obs)
```

```python
rd=[]
ffts=[]
bins=[]
ffts_sums=[]
rd_np=[]

for obs in obsl:
    rd.append(obs['raw_data'])
    ffts.append(obs['fft'])
    ffts_sums.append(np.sum(obs['fft']))
    bins.append(obs['fbins'])
    rd_np.append(obs['raw_data'][:7])
```
